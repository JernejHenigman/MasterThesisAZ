```python
import pandas as pd
import os
from datetime import datetime
```

```python
cwd = os.getcwd()
```

```python
cwd
```

```python
games = os.listdir(cwd+"/games")
```

```python
i = 0
for file in games:
    gameI = open(cwd+"/games/"+file,"r+")
    gamestats = file.split("=")
    #print(gamestats)
    iteration = gamestats[2].split(".")[0].split("_")[1]
    filetime = int(os.path.getctime(cwd+"/games/"+file))
    time = datetime.fromtimestamp(filetime)
    date = str(time).split(" ")[0]
    

    allGames = gameI.read()
    
    print(file)
    #print(i)
    #print(allGames)
    game = allGames.split("\n\n\n\n")
    matchlength = len(game)
    for gameNum in range(0,len(game)):
        filename = date+"_ckeckpoint="+str(iteration)+"_game="+str(gameNum)+".pgn"

        gametxt = game[gameNum]
        gametxt = gametxt.replace('[Event "?"]','[Event "AlphaZeroVSStockfish"]')
        gametxt = gametxt.replace('[Date "????.??.??"]','[Date "'+date+'"]')
        gametxt = gametxt.replace('[Round "?"]','[Round "'+str(gameNum)+'"]')
        gametxt = gametxt.replace('[Site "?"]','[Site "LocalTesting"]')
        if gameNum >= int(matchlength/2):
            gametxt = gametxt.replace('[White "?"]','[White "Stockfish"]')
            gametxt = gametxt.replace('[Black "?"]','[Black "AlphaZero"]')
        else:
            gametxt = gametxt.replace('[White "?"]','[White "AlphaZero"]')
            gametxt = gametxt.replace('[Black "?"]','[Black "Stockfish"]')
        newpgngame = open(cwd+"/pgnGames/"+filename,"w+")
        newpgngame.write(gametxt)
        newpgngame.close()
    gameI.close()
    i = i + 1
```
